```python
import numpy as np 
import pandas as pd
from pandas import DataFrame
ti = pd.read_csv(r'F:\pandas_data\titanic_train.csv')
print(ti.head(4))
print(DataFrame(ti))#NaN 为缺失值     
print(ti.columns)
print(ti.shape)
```

```python
age = ti['age']
print(age.head(19))
age_null = pd.isnull(age)#找出缺失值
print(age_null)
age_null_true = age[age_null]
print(age_null_true)
count = len(age_null_true)
print(count)
```

```python
mean_age = sum(ti['age'])/len(ti['age'])
print(mean_age)#存在一个缺失值即没有结果
good_age = ti['age'][age_null == False]
#print(good_age)
good_mean = sum(good_age)/len(good_age)
print(good_mean)
```

```python
#可直接调用pandas的mean函数
print(ti['age'].mean())
```

```python
#平均价格
clas=[1,2,3]
f_clas= {}
for i in clas :
    pclas = ti[ti['pclass']==i]
    fares = pclas['fare']
    fclas = fares.mean()
    f_clas[i]=fclas
print(f_clas)
    
```

```python
passen_sur = ti.pivot_table(index='pclass',values='survived',aggfunc=np.mean)
print(passen_sur)
```

```python
passen_age = ti.pivot_table(index='pclass',values='age')#上步已经指定function就会默认
print(passen_age)
```

```python
port_stats = ti.pivot_table(index='embarked',values=['fare','survived'],aggfunc=np.sum)
print(port_stats)
```

```python
def not_null_count(column):
    column_null = pd.isnull(column)
    null = column[column_null]
    return len(null)
column_null_count = ti.apply(not_null_count)
print(column_null_count)
```

```python
def which_class(row):
    pclass = row['pclass']
    if pd.isnull(pclass):
        return 'unknnown hhh'
    elif pclass == 1:
        return 'first class'
    elif pclass == 2:
        return 'second class'
    elif pclass ==3:
        return 'poor class'
classes = ti.apply(which_class,axis = 1)
print(classes)
        
```

```python
def age_class(row):
    age = row['age']
    if pd.isnull(age):
        return 'unknown'
    elif age < 18:
        return 'teen'
    elif age >=18:
        return 'older'
ages = ti.apply(age_class,axis = 1)
print(ages)
```

```python
ti['age_label']=ages
age_surve = ti.pivot_table(index = 'age_label',values = 'survived',aggfunc = np.mean)
print(age_surve)
```
